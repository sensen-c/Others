```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\sensenc\desktop\AIVI164.csv')[['Urls','Tag1','Tag2','Original','Order']]
dfa =  df[(df.Tag1 == 'Online')& (df.Order <0.79)]

y_true=dfa['Tag2'].to_list()
y_pred=dfa['Original'].to_list()

catlist = dfa.Original.drop_duplicates().to_list()
print(catlist)

df1 = dfa[dfa.Original != dfa.Tag2]
df1.head()

table = pd.pivot_table(df1,values='Urls',index=['Original','Tag2'],aggfunc='count')
tablhh = table[table.Urls >2]
tablehhh = tablhh.reset_index()
tableh = tablehhh.sort_values(['Original','Tag2'])
tableb = table.reset_index(0).reset_index(drop=False)
tablea = tableb[tableb.Urls > 2]
print(tableh)
```

```python
ax,fig = plt.subplots(50,20)

ax=sns.barplot(data = tableh, x = 'Original', y = 'Urls', hue = 'Tag2',palette='husl')

def change_width(ax, new_value) :
    locs = ax.get_xticks()
    for i,patch in enumerate(ax.patches):
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(locs[i//4] - (new_value * .5))

change_width(ax, .35)
plt.show()
```
